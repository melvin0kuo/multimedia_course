## <font color="Red">1.1 Calculate gray image from color image</font>
<img src="Chapter01_1.png" width="600px"> 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

imgBGR = cv2.imread("Commons.jpeg", cv2.IMREAD_COLOR)
img = imgBGR[:,:,::-1]

# Image size and informstion
nr, nc = img.shape[:2]
gray = np.zeros((nr,nc),dtype=np.uint8)
gray2 = np.zeros((nr,nc),dtype=np.uint8)
#-------------- Code Here ------------------#
for i in range(nr):
    for j in range(nc):
        gray[i,j] = img[i,j,0]*0.299 + img[i,j,1]*0.587 + img[i,j,2]*0.114
        gray2[i,j] = (img[i,j,0] + img[i,j,1] + img[i,j,2])/3



# Y = 0.299*img[:,:,0] + 0.587*img[:,:,1] + 0.114*img[:,:,2]
# Cb = 128 - 0.169*img[:,:,0] - 0.331*img[:,:,1] + 0.500*img[:,:,2]
# Cr = 128 + 0.500*img[:,:,0] - 0.419*img[:,:,1] - 0.081*img[:,:,2]

# Y = np.clip(Y,0,255).astype(np.uint8)
# Cb = np.clip(Cb,0,255).astype(np.uint8)
# Cr = np.clip(Cr,0,255).astype(np.uint8)

# gray = cv2.merge([Y,Cb,Cr])

#-------------- Stop Here ------------------#

plt.figure(figsize=(12, 10))
plt.subplot(1,3,1)   
plt.title('Image color') 
plt.imshow(img)
plt.subplot(1,3,2)     
plt.title('Image gray') 
plt.imshow(gray, cmap=plt.cm.gray)        
plt.subplot(1,3,3)     
plt.title('Image gray2') 
plt.imshow(gray2, cmap=plt.cm.gray)        
```

## <font color="Red">1.2 Calculate Cb and Cr</font>

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

imgBGR = cv2.imread("Commons.jpeg", cv2.IMREAD_COLOR)
img = imgBGR[:,:,::-1]

# Image size and informstion
nr, nc = img.shape[:2]
Cb = np.zeros((nr,nc),dtype=np.uint8)
Cr = np.zeros((nr,nc),dtype=np.uint8)
#-------------- Code Here ------------------#

for i in range(nr):
    for j in range(nc):
        Cb[i,j] = 128 - 0.169*img[i,j,0] - 0.331*img[i,j,1] + 0.500*img[i,j,2]
        Cr[i,j] = 128 + 0.500*img[i,j,0] - 0.419*img[i,j,1] - 0.081*img[i,j,2]

# Cb = 128 - 0.169*img[:,:,0] - 0.331*img[:,:,1] + 0.500*img[:,:,2]
# Cr = 128 + 0.500*img[:,:,0] - 0.419*img[:,:,1] - 0.081*img[:,:,2]
#-------------- Stop Here ------------------#
plt.figure(figsize=(12, 10))
plt.subplot(1,3,1)   
plt.title('Image color') 
plt.imshow(img)
plt.subplot(1,3,2)     
plt.title('Image Cb') 
plt.imshow(Cb) 
plt.subplot(1,3,3)     
plt.title('Image Cr') 
plt.imshow(Cr)        
```

```python
# Homework 2-2
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = np.zeros( [ 512, 512, 3 ], dtype = 'uint8' )

#-------------- Code Here ------------------#
for i in range(512):
    for j in range(512):
        if (i // 64 + j // 64) % 4 == 0:
            img[i, j] = [255, 0, 0]  # 紅色
        elif (i // 64 + j // 64) % 4 == 1:
            img[i, j] = [0, 255, 0]  # 綠色
        elif (i // 64 + j // 64) % 4 == 2:
            img[i, j] = [0, 0, 255]  # 藍色
        else:
            img[i, j] = [255, 255, 255]  # 白色


#-------------- Stop Here ------------------#        

plt.figure(figsize=(12, 10)) 
plt.title('original image') 
plt.imshow(img,cmap=plt.cm.gray)
```

## <font color="Red"> 1.7 Please create the following two images by Python</font><p>
Please use the following bit extraction method to acquire the bit plane image: <p>
val = 63 <p>
bit_plane = 0b00000001 <p>
val & bit_plane = 1 <p>
<img src="Chapter01_h3.png" width="800px">    

```python
val = 63 
bit_plane = 0b100000000
val_bit = val & bit_plane
print( "bit_plane = ", bit_plane )
print("val_bit = ", val_bit)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def bit_plane(f, bit):
#-------------- Code Here ------------------#    
    bit_plane_mask = 1 << bit
    bp = (f & bit_plane_mask) >> bit
    bp = bp * 255 

#-------------- Stop Here ------------------#            
    return bp    

img = cv2.imread("lenna.bmp", -1)
bit = 7
bitplane1 =  bit_plane(img, bit)   
    
plt.figure(figsize=(12, 10))
plt.subplot(2,2,1)   
plt.title('source image') 
plt.imshow(img,cmap=plt.cm.gray)
plt.subplot(2,2,2)     
plt.title('image of bit plane '+ str(bit)) 
plt.imshow(bitplane1,cmap=plt.cm.gray)
```
